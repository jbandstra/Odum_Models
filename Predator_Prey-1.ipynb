```python
import math as m
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
K1, K2, K3, K4, K5= 0.002, 0.05, 0.1, 0.03, 0.3
C0 = 5
H0 = 10
```

```python
def production(t, Q):
    C,H = Q
    E = 700
    dH = K1*E*H - K3*H*C - K2*H
    dC = K4*H*C - K5*C
    return [dC, dH]
```

```python
soln = solve_ivp(production, (0,100), [C0,H0],dense_output=True, method='RK45')
```

```python
t_arr = np.linspace(0,100,100)
plt.plot(t_arr, soln.sol(t_arr).T)
plt.xlabel("Time")
plt.ylabel("Count")
```
