```python
#Import statements
import math as m
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
#daily production and consumption

def StoredProd(t, Q):
    Mt = 1
    K = 0.9
    K1 = 0.2
    K2 = 0.04
    K3 = 0.01
    J = max (0,40*m.sin(t/15.9))
    
    M = Mt-K3*Q
    R = J/(1+K*M)
    
    return K1*R*M-K2*Q
soln = solve_ivp(StoredProd, (0,320), [5], dense_output = True)
print(soln)
t_arr = np.linspace(0,320, 100)
plt.plot(t_arr, soln.sol(t_arr).T)
plt.xlabel ('Time')
plt.ylabel('Stored Product')
    
plt.figure()
plt.plot(t_arr, 1-0.01*soln.sol(t_arr).T)
plt.xlabel ('Time')
plt.ylabel('Materials Available')
```
